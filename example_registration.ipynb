```python
import obi_auth
from entitysdk import Client

from obi_notebook import get_projects

environment_str = "staging"

token = obi_auth.get_token(environment=environment_str)
project_context = get_projects.get_projects(token, env=environment_str)
```

```python
from obi_notebook import register_result

client = Client(environment=environment_str, project_context=project_context, token_manager=token)
register_result.register(client)
```

```python
from entitysdk import models

res = client.search_entity(entity_type=models.AnalysisNotebookResult).all()
for res_ in res:
    print((res_.name, res_.description))
```
